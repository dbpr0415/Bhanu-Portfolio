```
import time
import glob
from pathlib import Path

# Choose device
device = 'cpu'  # or 'cuda' if using GPU

# Select test images (first 10–20 is enough)
image_paths = glob.glob('/content/dataset/images/val/*.jpg')[:10]
weights_path = '/content/yolov7/runs/train/yolov7-custom2/weights/best.pt'
output_dir = 'inference_temp/'
Path(output_dir).mkdir(exist_ok=True)

times = []

for img_path in image_paths:
    start = time.time()

    # WRAP img_path in quotes to avoid errors
    !python detect.py --weights "$weights_path" \
                      --img 640 \
                      --conf 0.25 \
                      --source "$img_path" \
                      --device $device \
                      --save-txt --nosave \
                      --project $output_dir --name test_run --exist-ok

    end = time.time()
    times.append(end - start)

avg = sum(times) / len(times)
fps = 1 / avg

print(f"\n✅ Avg Inference Time per image on {device.upper()}: {avg:.4f} sec")
print(f"🎯 FPS: {fps:.2f}")
```
